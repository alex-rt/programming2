```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

wk1 = pd.read_csv('wktime.T001.his', delimiter=' ',header=None) 
wk2 = pd.read_csv('wktime.T004.his', delimiter=' ',header=None) 

plt.plot(wk1[0], wk1[1], color='mediumpurple', label='time=1') 
plt.plot(wk2[0], wk2[1], color='lightseagreen', label='time=4') 
plt.legend(loc='best')
plt.xlabel("wealth")
plt.ylabel("probability")
plt.title("probability distribution")
plt.xscale("log")                      
plt.yscale("log")                      
plt.xlim((0.000001, 10.0))           
plt.ylim((0.000001, 100000.0))  
plt.tight_layout()
```

```python
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)

plt.plot(x, np.sin(x - 0), color='blue',label='sin(x)')       
plt.plot(x, np.sin(x - 1), color='g',label='sin(x-1)')         
plt.plot(x, np.sin(x - 2), color='0.75',label='sin(x-2)')       
plt.plot(x, np.sin(x - 3), color='#FFDD44',label='sin(x-3)')    
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3),label='sin(x-4)') 
plt.plot(x, np.sin(x - 5), color='chartreuse',label='sin(x-5)'); 
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Trigonometric graph (sin)")
plt.legend(loc='best')
```
